### Preprocessing of lyrics

* remove empty lines
* remove words and characters that should not be included in the lyrics
* remove lyrics that are not correctly formatted


```python
df = pd.read_csv("spotify_songs_with_lyrics.csv")

df_dropped = df[["playlist_genre", "lyrics"]]

# Define functions for the required operations

def remove_empty_lyrics(df):
    """Remove all rows with empty lyrics."""
    return df[df["lyrics"].notna() & (df["lyrics"] != "")]


def remove_first_line_lyrics(df):
    """Remove the first line in the lyrics."""
    df["lyrics"] = df["lyrics"].apply(
        lambda x: "\n".join(x.split("\n")[1:]) if x else x
    )
    return df


def remove_empty_lines_lyrics(df):
    """Remove empty lines in the lyrics."""
    df["lyrics"] = df["lyrics"].apply(
        lambda x: "\n".join(filter(None, x.split("\n"))) if x else x
    )
    return df


def remove_lines_starting_with_brackets(df):
    """Remove all lines in the lyrics that start with '['."""
    df["lyrics"] = df["lyrics"].apply(
        lambda x: "\n".join(line for line in x.split("\n") if not line.startswith("["))
        if x
        else x
    )
    return df


def remove_embed_at_end(df):
    """Remove 'Embed' at the end of each lyric."""
    df["lyrics"] = df["lyrics"].apply(
        lambda x: x.replace("Embed", "").strip() if x else x
    )
    return df


# Apply the operations
df_dropped = remove_empty_lyrics(df_dropped)
df_dropped = remove_first_line_lyrics(df_dropped)
df_dropped = remove_lines_starting_with_brackets(df_dropped)
# df_dropped = remove_empty_lines_lyrics(df_dropped)
df_dropped = remove_embed_at_end(df_dropped)

# Display the first few rows of the cleaned dataframe
df_dropped
```

```python
df_dropped.to_csv("genreXlyrics.csv", index=False)
```

```python
import re

data_reloaded = df_dropped.copy()


def is_exclusively_song_titles_refined(lyrics):
    # Check if the lyrics is not a string (e.g., NaN), then it's not exclusively song titles
    if not isinstance(lyrics, str):
        return False

    # Split the lyrics into lines and remove empty lines
    lines = [line.strip() for line in lyrics.split("\n") if line.strip()]

    # Regular expression to match the pattern "Artist - 'Song Title'" and variations
    title_pattern = re.compile(r'^.+ - ".+".*$')

    # Check each line, if any line does not match the title pattern or is a date, return False (keep the row)
    for line in lines:
        if title_pattern.match(line) or re.match(r"^\d+/\d+", line):
            return True
    return False


# Apply this refined function and filter the data to remove rows that are exclusively song titles or similar patterns
exclusively_titles_filtered_data = data_reloaded[
    ~data_reloaded["lyrics"].apply(is_exclusively_song_titles_refined)
]

# Display the first few rows of the dataset after filtering
print(exclusively_titles_filtered_data)
len(exclusively_titles_filtered_data) - len(data_reloaded)
```

### Split lyrics in stanzas

```python
# Split the lyrics at every empty line to increase the number of training data

df_split = exclusively_titles_filtered_data.copy()

# Function to split lyrics at empty lines
def split_lyrics(lyrics):
    split_lyrics = lyrics.split("\n\n")  # Splitting at empty lines
    return [
        lyric.strip() for lyric in split_lyrics if lyric.strip()
    ]  # Removing any leading/trailing spaces

# Applying the function to each row in the lyrics column
df_split["lyrics"] = df_split["lyrics"].apply(split_lyrics)

# remove duplicates
df_split = df_split.drop_duplicates(subset=["lyrics"])

# Exploding the dataframe so each piece of lyrics is in a separate row
df_split = df_split.explode("lyrics").reset_index(drop=True)

# remove '\n' from lyrics
df_split["lyrics"] = df_split["lyrics"].apply(lambda x: str(x).replace("\n", ". "))

df_split
```

### Equalizing the Number of Songs per Genre through Downsampling

```python
# count each genre
print(df_split["playlist_genre"].value_counts())

# reduce each genre to the same amount of songs
df_split = df_split.groupby("playlist_genre").head(
    df_split["playlist_genre"].value_counts().min()
)

# count each genre
print(df_split["playlist_genre"].value_counts())
df_split
```

```python
# pop = 0, rap = 1, rock = 2, rnb = 3, latin = 4, edm = 5
df_split["playlist_genre"] = df_split["playlist_genre"].replace(
    ["pop", "rap", "rock", "r&b", "latin", "edm"], [0, 1, 2, 3, 4, 5]
)
df_split
```

```python
df_split.to_csv("data.csv", index=False)
```
