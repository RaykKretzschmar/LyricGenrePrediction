```python
df = pd.read_csv('spotify_songs_with_lyrics.csv')

df_dropped = df[['playlist_genre', 'lyrics']]

# Define functions for the required operations

def remove_empty_lyrics(df):
    """Remove all rows with empty lyrics."""
    return df[df['lyrics'].notna() & (df['lyrics'] != '')]

def remove_first_line_lyrics(df):
    """Remove the first line in the lyrics."""
    df['lyrics'] = df['lyrics'].apply(lambda x: '\n'.join(x.split('\n')[1:]) if x else x)
    return df

def remove_empty_lines_lyrics(df):
    """Remove empty lines in the lyrics."""
    df['lyrics'] = df['lyrics'].apply(lambda x: '\n'.join(filter(None, x.split('\n'))) if x else x)
    return df

def remove_lines_starting_with_brackets(df):
    """Remove all lines in the lyrics that start with '['."""
    df['lyrics'] = df['lyrics'].apply(lambda x: '\n'.join(line for line in x.split('\n') if not line.startswith('[')) if x else x)
    return df

def remove_embed_at_end(df):
    """Remove 'Embed' at the end of each lyric."""
    df['lyrics'] = df['lyrics'].apply(lambda x: x.replace('Embed', '').strip() if x else x)
    return df

# Apply the operations
df_dropped = remove_empty_lyrics(df_dropped)
df_dropped = remove_first_line_lyrics(df_dropped)
df_dropped = remove_lines_starting_with_brackets(df_dropped)
# df_dropped = remove_empty_lines_lyrics(df_dropped)
df_dropped = remove_embed_at_end(df_dropped)

# Display the first few rows of the cleaned dataframe
df_dropped
```

```python
# Split the lyrics at every empty line to increase the number of training data

# Function to split lyrics at empty lines
def split_lyrics(lyrics):
    split_lyrics = lyrics.split('\n\n')  # Splitting at empty lines
    return [lyric.strip() for lyric in split_lyrics if lyric.strip()]  # Removing any leading/trailing spaces

# Applying the function to each row in the lyrics column
df_split = df_dropped.copy()
df_split['lyrics'] = df_split['lyrics'].apply(split_lyrics)

# Exploding the dataframe so each piece of lyrics is in a separate row
df_split = df_split.explode('lyrics').reset_index(drop=True)

# Display the first few rows of the expanded dataframe
df_split
```

```python
# train a machine learning model to predict the genre of a song based on its lyrics with pytorch
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torchtext import data
from torchtext import datasets
import random
import spacy
import time
```

```python
# set random seeds to make sure results are reproducible
SEED = 1234
torch.manual_seed(SEED)
torch.backends.cudnn.deterministic = True

# set device to GPU if available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)

# load spacy tokenizer
spacy_en = spacy.load('en_core_web_sm')
```

```python
# define tokenizer function
def tokenizer(text): # create a tokenizer function
    return [tok.text for tok in spacy_en.tokenizer(text)]
```

```python

# define the fields of the dataset
TEXT = data.Field(tokenize=tokenizer)
LABEL = data.LabelField(dtype=torch.float)

# load the dataset
fields = [('lyrics', TEXT), ('playlist_genre', LABEL)]
training_data = data.TabularDataset(path='spotify_songs_with_lyrics.csv', format='csv', fields=fields, skip_header=True)

# split the dataset into training, validation and test set
train_data, valid_data, test_data = training_data.split(split_ratio=[0.7, 0.2, 0.1], random_state=random.seed(SEED))

# build the vocabulary
TEXT.build_vocab(train_data, vectors="glove.6B.100d")
LABEL.build_vocab(train_data)

# set batch size
BATCH_SIZE = 64
```

```python

# create iterators for the dataset
train_iterator, valid_iterator, test_iterator = data.BucketIterator.splits(
    (train_data, valid_data, test_data), 
    batch_size=BATCH_SIZE,
    sort_key=lambda x: len(x.lyrics),
    device=device)
```

```python

# define the model
class RNN(nn.Module):
    def __init__(self, input_dim, embedding_dim, hidden_dim, output_dim):
        super().__init__()
        
        self.embedding = nn.Embedding(input_dim, embedding_dim)
        self.rnn = nn.RNN(embedding_dim, hidden_dim)
        self.fc = nn.Linear(hidden_dim, output_dim)
        
    def forward(self, text):
        embedded = self.embedding(text)
        output, hidden = self.rnn(embedded)
        assert torch.equal(output[-1,:,:], hidden.squeeze(0))
        return self.fc(hidden.squeeze(0))
    
```

```python

# define hyperparameters
INPUT_DIM = len(TEXT.vocab)
EMBEDDING_DIM = 100
HIDDEN_DIM = 256
OUTPUT_DIM = 5

# create the model
model = RNN(INPUT_DIM, EMBEDDING_DIM, HIDDEN_DIM, OUTPUT_DIM)

# define optimizer and loss function
optimizer = optim.Adam(model.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

# push model to GPU if available
model = model.to(device)
criterion = criterion.to(device)
```

```python

# define accuracy function
def accuracy(preds, y):
    preds = preds.argmax(dim=1, keepdim=True)
    correct = preds.eq(y.view_as(preds)).sum()
    acc = correct.float()/preds.shape[0]
    return acc
```

```python

# define training function
def train(model, iterator, optimizer, criterion):
    
    epoch_loss = 0
    epoch_acc = 0
    
    model.train()
    
    for batch in iterator:
        
        optimizer.zero_grad()
        
        predictions = model(batch.lyrics).squeeze(1)
        
        loss = criterion(predictions, batch.playlist_genre.long())
        
        acc = accuracy(predictions, batch.playlist_genre.long())
        
        loss.backward()
        
        optimizer.step()
        
        epoch_loss += loss.item()
        epoch_acc += acc.item()
        
    return epoch_loss/len(iterator), epoch_acc/len(iterator)

# define evaluation function
```
